```python
dc = df.copy()
```

## Dropping columns with one unique value

```python
dc.loc[:,dc.nunique()==1].columns
```

```python
nd = dc.loc[:,dc.nunique()>1]
nd.head()
```

## Separating DataFrames based on 'headers_x_github_event' column

```python
splt = {x:y for x, y in nd.groupby('headers_x_github_event', as_index=False)}
```

```python
for i,j in zip(splt.keys(),[splt[x].shape for x in splt.keys()]):
    print(i,j)
```

```python
spdf = {x:splt[x].dropna(how='all',axis=1) for x in splt.keys()}
for i,j in zip(spdf.keys(),[spdf[x].shape for x in spdf.keys()]):
    print(i,j)
```

```python
spdf = {x:spdf[x].loc[:,spdf[x].nunique()>1] for x in spdf.keys()}
for i,j in zip(spdf.keys(),[spdf[x].shape for x in spdf.keys()]):
    print(i,j)
```
